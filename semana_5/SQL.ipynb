# SQL
O que é o SQL? Da forma mais simples possível, é a linguagem usada para consulta e manipulação de banco de dados. Aqui deixarei uma "cheat sheet" dos comandos de SQL mais usados.

### Criação de Tabela
```sql

CREATE TABLE base_compras (
    Id int,
    Nome varchar(50),
    Sobrenome varchar(50),
    Estado varchar(2),
    Gastos decimal,
    Data_Compra date
);

```

### Colocando valores na tabela

```sql

NSERT INTO base_compras
VALUES
    (152, 'Andre', 'Silva', 'MG', 351.50, '2018-01-22'),
    (222, 'Barbara', 'Toledo', 'SP', 250.10, '2018-05-15'),
    (451, 'Carlos', 'Pinheiro', 'MG', 455.00, '2017-02-05'),
    (754, 'Eduardo', 'Silva', 'SP', 390.10, '2018-04-10'),
    (897, 'Juliana', 'Oliveira', 'MG', 150.50, '2017-03-01'),
    (852, 'Maria', 'Lima', 'MG', 325.90, '2018-05-30'),
    (997, 'Ricardo', 'Pereira', 'MG', 332.59, '2018-05-25'),
    (535, 'Vanessa', 'Costa', 'SP', 241.57, '2017-04-30');

```

### Puxando dados da tabela

```sql

SELECT * from base_compras

```

### Filtrando valores da tabela

```sql

SELECT * from base_compras
WHERE Estado = "SP"

```

### Salvando uma tabela 

```sql

CREATE TABLE clientes_SP AS 
SELECT * from base_compras
WHERE Estado = "SP";

```
